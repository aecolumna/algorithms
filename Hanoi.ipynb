```python
def printMove(fr,to):
    print("Move from " + str(fr) + " to " + str(to))
```

```python
def Towers(n,fr,to,spare):
 
        if n == 1:
            printMove(fr,to)
        else:
            Towers(n-1,fr, spare, to)
            Towers(1, fr,to,spare)
            Towers(n-1,spare, to,fr)
```

```python
Towers(10,1,3,2)
```

```python
Towers(4,1,3,2)
```
